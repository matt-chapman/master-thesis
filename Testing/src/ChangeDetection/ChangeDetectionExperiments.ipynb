```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(seed=111111)
np.set_printoptions(precision=2, suppress=True)
```

```python
signal = np.random.randint(10, size=20)
print "Signal: {}".format(signal)
plt.plot(range(0, 20), signal)
plt.show()
```

```python
signal2 = np.random.normal(scale=1, size=200)
signal2[:50] +=50; signal2[50:] += 55; signal2[150:] -= 20
signal2[110] = 0
plt.plot(signal2)
plt.ylim(0,100)
plt.show()
```
